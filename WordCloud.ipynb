```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
my_sentence =  "Come and Explore the world of Python, Data Science, Machine learning and Artificial Intelligence with me"
```

```python
wordcloud = WordCloud(width = 1000, height = 500, background_color = "grey").generate(my_sentence)
```

```python
plt.figure(figsize= (15,10))
plt.imshow(wordcloud, interpolation = "bilinear")
plt.axis('off')
```
