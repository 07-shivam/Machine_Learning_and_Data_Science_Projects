```python
corr_mat = data.corr()
```

```python
for row in range (len(corr_mat)):
    for col in range(row):
        if corr_mat.iloc[row][col]<-0.5 or corr_mat.iloc[row][col]>0.5:
            print("*"*5)
            print(row,col)
            print(corr_mat.iloc[row][col])
```

```python
for row in range (len(corr_mat)):
    for col in range(row):
        if corr_mat.iloc[row][col]<-0.5 or corr_mat.iloc[row][col]>0.5:
            print("*"*5)
            print(row,col)
            print(corr_mat.columns[row])
```

```python
drop_columns = []
for row in range (len(corr_mat)):
    for col in range(row):
        if corr_mat.iloc[row][col]<-0.5 or corr_mat.iloc[row][col]>0.5:
            print("*"*5)
            print(corr_mat.columns[row])
            drop_columns.append(corr_mat.columns[row])
```

```python
## Droping Highly Corelated columns
```

```python
data.drop(set(drop_columns), axis = 1, inplace = True)
```
