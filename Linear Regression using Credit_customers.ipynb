```python
import pandas as pd
```

```python
data = pd.read_csv("credit_customers (1).csv")
```

```python
data.head(10)
```

```python
data.isnull().sum()
```

```python
for i in data.columns:
    try:
        print(f"""column is {i} and number of None is {data[i].value_counts()["none"]} is None""")
    except :
        print("no None")
```

```python
data.drop(["other_parties","other_payment_plans","own_telephone","checking_status"], axis = 1, inplace = True)
```

```python
data
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
Data_encode = []
for i, dtype in data.dtypes.items():
    if dtype != int and  dtype != float:
        Data_encode.append(i)
```

```python
Data_encode
```

```python
for i in Data_encode:
    data[i] = le.fit_transform(data[i])
```

```python
data
```

```python
data.columns
```

```python
X = data[['duration', 'credit_history', 'purpose', 'credit_amount',
       'savings_status', 'employment', 'installment_commitment',
       'personal_status', 'residence_since', 'property_magnitude', 'age',
       'housing', 'existing_credits', 'job', 'num_dependents',
       'foreign_worker']]
```

```python
y = data["class"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logR = LogisticRegression()
```

```python
logR.fit(X_train,y_train)
```

```python
ypred = logR.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(ypred,y_test)
```

```python
logR2 = LogisticRegression(solver = "liblinear")
logR2.fit(X_train,y_train)
ypred = logR2.predict(X_test)
accuracy_score(ypred,y_test)
```

```python
logR3 = LogisticRegression(solver = "sag")
logR3.fit(X_train,y_train)
ypred = logR3.predict(X_test)
accuracy_score(ypred,y_test)
```

```python
logR4 = LogisticRegression(solver = "saga")
logR4.fit(X_train,y_train)
ypred = logR4.predict(X_test)
accuracy_score(ypred,y_test)
```

```python
logR5 = LogisticRegression(solver = "newton-cg")
logR5.fit(X_train,y_train)
ypred = logR5.predict(X_test)
accuracy_score(ypred,y_test)
```
